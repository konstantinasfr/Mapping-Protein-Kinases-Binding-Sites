```python
import pickle
import subprocess as sp
import os 
import re
import numpy as np
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from collections import Counter
```

```python
a_file = open("data.pkl", "rb")
output = pickle.load(a_file)
print(output)
a_file.close()
```

```python
new_dict1 = dict([(key, val) for key, val in 
           output.items() if len(val) > 10])
```

```python
# 1l3i_3
def find_neibs(pocket):
    counter = 0
    similarities = []
    for key in new_dict1:
        for value in new_dict1[key]:
            counter += 1
            print(str(counter) +" "+ str(key)+"   "+str(value),end="\r")
            command = "./find_similarities1.sh convertions30/"+str(pocket)+" scPDB_sites_converted/"+str(value)+".pdb"
    #        print(sp.getoutput(command))
            try:
                similarities.append([key,value,float(sp.getoutput(command))])
            except ValueError as ve:
                print(value)
    return similarities 
```
