```python
hxd_dic = {}
with open('HXD_sites.txt','r') as f:
    for line in f:
        count = 0
        for word in line.split("||"):
            if count == 0:
                name = word[:4]
                hxd_dic[name] = []
            elif count == 1:
                hxd_dic[name].append(int(word[:3]))
            elif count == 2 and word != '\n':
#                 pos1 = word[:3]
                hxd_dic[name].append(int(word[:3]))
            elif count == 3 and word != '\n':
#                 pos1 = word[:3]
                hxd_dic[name].append(int(word[:3]))
            elif count == 4 and word != '\n':
#                 print(word[:3])
                hxd_dic[name].append(int(word[:3]))
                
            count += 1
```

```python
hxd_dic
```

```python
hxd_total = []

for pocket_file in total_pocks:
    positions1 = []
    with open("convertions30/"+str(pocket_file),'r') as f:
        for line in f:
            j = 2
            sentence = []
            while ( "." not in line.split()[j]):
                sentence.append(line.split()[j])
                j += 1

            positions1.append(re.sub("[^0-9]", "",sentence[-1]))
            
            
            
            
    positions1 = [int(i) for i in positions1]   

    if pocket_file[:4] in hxd_dic:
        p = str(pocket_file[:4])+"_"+re.findall(r'[0-9]+', pocket_file[4:])[0]
#         print(p)
        if len(hxd_dic[pocket_file[:4]]) == 1:
            cond1 = hxd_dic[pocket_file[:4]][0] in positions1 and hxd_dic[pocket_file[:4]][0]+2 in positions1
            if cond1:
                hxd_total.append([p,"hxd"])
            else:
                hxd_total.append([p,"non_hxd"])

        if len(hxd_dic[pocket_file[:4]]) == 2:
            cond1 = hxd_dic[pocket_file[:4]][0] in positions1 and hxd_dic[pocket_file[:4]][0]+2 in positions1
            cond2 = hxd_dic[pocket_file[:4]][1] in positions1 and hxd_dic[pocket_file[:4]][1]+2 in positions1
            if cond1 or cond2:
                hxd_total.append([p,"hxd"])
            else:
                hxd_total.append([p,"non_hxd"])
                
        if len(hxd_dic[pocket_file[:4]]) == 3:
            cond1 = hxd_dic[pocket_file[:4]][0] in positions1 and hxd_dic[pocket_file[:4]][0]+2 in positions1
            cond2 = hxd_dic[pocket_file[:4]][1] in positions1 and hxd_dic[pocket_file[:4]][1]+2 in positions1
            cond3 = hxd_dic[pocket_file[:4]][2] in positions1 and hxd_dic[pocket_file[:4]][2]+2 in positions1
            if cond1 or cond2 or cond3:
                hxd_total.append([p,"hxd"])
            else:
                hxd_total.append([p,"non_hxd"])
                
        if len(hxd_dic[pocket_file[:4]]) == 4:
            cond1 = hxd_dic[pocket_file[:4]][0] in positions1 and hxd_dic[pocket_file[:4]][0]+2 in positions1
            cond2 = hxd_dic[pocket_file[:4]][1] in positions1 and hxd_dic[pocket_file[:4]][1]+2 in positions1
            cond3 = hxd_dic[pocket_file[:4]][2] in positions1 and hxd_dic[pocket_file[:4]][2]+2 in positions1
            cond4 = hxd_dic[pocket_file[:4]][3] in positions1 and hxd_dic[pocket_file[:4]][3]+2 in positions1
            if cond1 or cond2 or cond3 or cond4:
                hxd_total.append([p,"hxd"])
            else:
                hxd_total.append([p,"non_hxd"])
                
                
    else:
        p = str(pocket_file[:4])+"_"+re.findall(r'[0-9]+', pocket_file[4:])[0]
        hxd_total.append([p,"non_hxd"])
```

```python
counter = 0
wtf = []
for pocs in hxd_total:
    if pocs[1] == "hxd":
        counter += 1
        wtf.append(pocs[0][:4])
print(counter)
print(len(wtf))
print(len(np.unique(wtf)))
dic_wtf = Counter(wtf)
print(dic_wtf)
```
